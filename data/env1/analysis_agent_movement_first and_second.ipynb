# Analyse
Analyse all validate data

### First train
Train to find the best algorithm and learning rate

```python
names = []
values = []
path_folders = 'Comp3_movement_first'
for folder in os.listdir(path_folders):
    for filename in os.listdir(f'{path_folders}/{folder}'):
        if filename.endswith('.csv') and filename != 'historical.csv':
            data = pd.read_csv(f'{path_folders}/{folder}/{filename}', sep=';')
            names.append(f'{folder}-{filename}')
            
            data['is_win'] = (data['last_reward'] > 600)
            data_only_win = data[data.is_win]
            
            acc = data['is_win'].sum() / len(data)
            values.append([
                folder,
                acc,
                data_only_win.describe()['steps']['mean'],
                data_only_win.describe()['time']['mean'],
                data_only_win.describe()['mean_fps']['mean'],
                data_only_win.describe()['reward']['mean'],
                data_only_win.describe()['last_reward']['mean'],
                data_only_win.describe()['steps']['std'],
                data_only_win.describe()['time']['std'],
                data_only_win.describe()['mean_fps']['std'],
                data_only_win.describe()['reward']['std'],
                data_only_win.describe()['last_reward']['std']
            ])
```

```python
my_dataframe = pd.DataFrame()
my_dataframe = my_dataframe.append(values)
my_dataframe = my_dataframe.set_axis(names, axis=0)
my_dataframe = my_dataframe.set_axis(['folder', 'acc', 'steps_mean', 'time_mean', 'mean_fps_mean', 'reward_mean',
                                      'last_reward_mean', 'steps_std', 'time_std', 'mean_fps_std', 'reward_std',
                                      'last_reward_std'], axis=1)
```

```python
my_dataframe = my_dataframe.sort_values(by='folder')
```

```python
# All best checkpoints
my_dataframe.sort_values(by='acc', ascending=False)
```

### Second train
To find the best hiperparams

```python
names = []
values = []
path_folders = 'Comp3_movement_second'
for folder in os.listdir(path_folders):
    for filename in os.listdir(f'{path_folders}/{folder}'):
        if filename.endswith('.csv') and filename != 'historical.csv':
            data = pd.read_csv(f'{path_folders}/{folder}/{filename}', sep=';')
            names.append(f'{folder}-{filename}')
            
            data['is_win'] = (data['last_reward'] > 600)
            data_only_win = data[data.is_win]
            
            acc = data['is_win'].sum() / len(data)
            values.append([
                folder,
                acc,
                data_only_win.describe()['steps']['mean'],
                data_only_win.describe()['time']['mean'],
                data_only_win.describe()['mean_fps']['mean'],
                data_only_win.describe()['reward']['mean'],
                data_only_win.describe()['last_reward']['mean'],
                data_only_win.describe()['steps']['std'],
                data_only_win.describe()['time']['std'],
                data_only_win.describe()['mean_fps']['std'],
                data_only_win.describe()['reward']['std'],
                data_only_win.describe()['last_reward']['std']
            ])
```

```python
my_dataframe = pd.DataFrame()
my_dataframe = my_dataframe.append(values)
my_dataframe = my_dataframe.set_axis(names, axis=0)
my_dataframe = my_dataframe.set_axis(['folder', 'acc', 'steps_mean', 'time_mean', 'mean_fps_mean', 'reward_mean',
                                      'last_reward_mean', 'steps_std', 'time_std', 'mean_fps_std', 'reward_std',
                                      'last_reward_std'], axis=1)
```

```python
my_dataframe = my_dataframe.sort_values(by='folder')
```

```python
# All best checkpoints
my_dataframe[my_dataframe['acc'] == 1.0].sort_values(by='steps_mean', ascending=True)
```

```python
# All checkpoints
my_dataframe.sort_values(by=['acc', 'steps_mean'], ascending=[False, True])
```

```python
best_each_model = []
current_test = my_dataframe['folder'][0]
best_current_step = my_dataframe['steps_mean'][0]
best_current_acc = my_dataframe['acc'][0]
idx_best = 0
for idx, value in enumerate(my_dataframe.iterrows()):
    if current_test == value[1]['folder']:
        if value[1]['steps_mean'] < best_current_step and value[1]['acc'] >= best_current_acc or value[1]['acc'] > best_current_acc:
            idx_best = idx
            best_current_step = value[1]['steps_mean']
            best_current_acc = value[1]['acc']
    else:
        best_each_model.append(my_dataframe.iloc[idx_best])
        
        best_current_step = value[1]['steps_mean']
        best_current_acc = value[1]['acc']
        idx_best = idx
        current_test = value[1]['folder']
best_each_model.append(my_dataframe.iloc[idx_best])
```

```python
# Best checkpoint of each model
best_each_model_pd = pd.DataFrame(best_each_model)
best_each_model_pd.sort_values(by=['acc', 'steps_mean'], ascending=[False, True])
```
