```python
import os
import pandas as pd
from matplotlib import pyplot as plt
```

```python
pd.set_option('display.max_rows', 500)
```

### Tracker test

```python
names = []
values = []
path_folders = 'Comp3_tracker'
for folder in os.listdir(path_folders):
    for filename in os.listdir(f'{path_folders}/{folder}'):
        if filename.endswith('.csv') and filename != 'historical.csv':
            data = pd.read_csv(f'{path_folders}/{folder}/{filename}', sep=';')
            names.append(f'{folder}-{filename}')
            
            data['is_win'] = (data['last_reward'] > 600)
            data_only_win = data[data.is_win]
            
            acc = data['is_win'].sum() / len(data)
            values.append([
                folder,
                acc,
                data_only_win.describe()['steps']['mean'],
                data_only_win.describe()['time']['mean'],
                data_only_win.describe()['mean_fps']['mean'],
                data_only_win.describe()['reward']['mean'],
                data_only_win.describe()['last_reward']['mean'],
                data_only_win.describe()['steps']['std'],
                data_only_win.describe()['time']['std'],
                data_only_win.describe()['mean_fps']['std'],
                data_only_win.describe()['reward']['std'],
                data_only_win.describe()['last_reward']['std']
            ])
```

```python
my_dataframe = pd.DataFrame()
my_dataframe = my_dataframe.append(values)
my_dataframe = my_dataframe.set_axis(names, axis=0)
my_dataframe = my_dataframe.set_axis(['folder', 'acc', 'steps_mean', 'time_mean', 'mean_fps_mean', 'reward_mean',
                                      'last_reward_mean', 'steps_std', 'time_std', 'mean_fps_std', 'reward_std',
                                      'last_reward_std'], axis=1)
```

```python
my_dataframe = my_dataframe.sort_values(by='folder')
```
