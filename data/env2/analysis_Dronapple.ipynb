#### Apples Collected

```python
def print_result_collected(result, is_collected=True):
    print(f'Total: {len(result)}')
    if len(result) > 0:
        if is_collected:
            print(f'Acuracy collected without collision: {(len(result[result["last_code_arm_collision"] == 10000]["last_code_arm_collision"]) / len(result)) * 100}%')
            # Arm collision with BlackAlder (tree)
            print(f'Porcentege collected with 10001: {(len(result[result["last_code_arm_collision"] == 10001]["last_code_arm_collision"]) / len(result)) * 100}%')
            # Arm collision with tefcefala_LOD1_Blueprint (normal apple)
            print(f'Porcentege collected with 10002: {(len(result[result["last_code_arm_collision"] == 10002]["last_code_arm_collision"]) / len(result)) * 100}%')
            # Arm collision with tefcefala_LOD1_obstructed_Blueprint (obstructed apple)
            print(f'Porcentege collected with 10003: {(len(result[result["last_code_arm_collision"] == 10003]["last_code_arm_collision"]) / len(result)) * 100}%')
            print(result['last_code_arm_collision'].value_counts())
        else:
            # Arm collision with BlackAlder (tree)
            print(f'Porcentege not collected because 10001: {(len(result[result["last_code_arm_collision"] == 10001]["last_code_arm_collision"]) / len(result)) * 100}%')
            # Arm collision with tefcefala_LOD1_Blueprint (normal apple)
            print(f'Porcentege not collected because 10002: {(len(result[result["last_code_arm_collision"] == 10002]["last_code_arm_collision"]) / len(result)) * 100}%')
            # Arm collision with tefcefala_LOD1_obstructed_Blueprint (obstructed apple)
            print(f'Porcentege not collected because 10003: {(len(result[result["last_code_arm_collision"] == 10003]["last_code_arm_collision"]) / len(result)) * 100}%')
            print(result['last_code_arm_collision'].value_counts())
```

```python
result_NR = data_collected_filtered[data_collected_filtered["last_code_grab_overlap"].between(21000, 21999)]
result_NG = data_collected_filtered[data_collected_filtered["last_code_grab_overlap"].between(22000, 22999)]
result_OR = data_collected_filtered[data_collected_filtered["last_code_grab_overlap"].between(23000, 23999)]
result_OG = data_collected_filtered[data_collected_filtered["last_code_grab_overlap"].between(24000, 24999)]

print('#### NR ####')
print_result_collected(result_NR)
print('#### NG ####')
print_result_collected(result_NG)
print('#### OR ####')
print_result_collected(result_OR)
print('#### OG ####')
print_result_collected(result_OG)
```

```python
data_collected_filtered2 = data_collected_filtered[data_collected_filtered['last_code_grab_overlap'] < 20000]
result_fail = data_collected_filtered2[data_collected_filtered2['last_code_arm_collision'] > 10000]
print_result_collected(result_fail, is_collected=False)
```

```python
# Porcentege of fails when tried to collect an apple
p = len(result_fail) / (len(data_collected_filtered[data_collected_filtered["last_code_grab_overlap"] >= 21000]) + len(result_fail))
print(f'Fail: {p * 100}%')
print(f'Success: {100 - p * 100}%')
```
